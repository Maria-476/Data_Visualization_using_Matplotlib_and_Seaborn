### 📊 What is Data Visualization?

Data visualization is the process of **representing data in graphical or visual formats** (like charts, graphs, maps) so it’s easier to **see patterns, trends, and insights** rather than just looking at raw numbers.


### 🔑 Important Python Libraries for Data Visualization

* **Matplotlib** – Foundation library for creating static, animated, and interactive plots.
* **Seaborn** – Built on Matplotlib; makes attractive and statistical plots with less code.
* **Plotly** – For interactive and web-based plots; great for dashboards.
* **Bokeh** – Interactive, web-friendly visualizations with high performance.
* **Altair** – Declarative syntax for clean and concise visualizations.
* **Pandas (built-in plotting)** – Quick plots directly from DataFrames.
* **ggplot (plotnine)** – Grammar of graphics style, similar to R’s ggplot2.

---

### **Grammaer of Graphics**

### 🎨 What is Grammar of Graphics?

The **Grammar of Graphics** is a framework for building visualizations.
Instead of memorizing “chart types,” it breaks a graph into **basic building blocks** (data, aesthetics, geometry, etc.). You combine these blocks like Lego pieces to create any chart.


### 🧩 Key Components of Grammar of Graphics

* **Data**

  * The dataset you want to visualize (e.g., sales, population, exam scores).

* **Aesthetic (aes)**

  * How data maps to visuals: x-axis, y-axis, color, size, shape.

* **Geometry (geom)**

  * The actual chart type (e.g., points for scatter, bars for bar chart, lines for line graph).

* **Facets**

  * Splitting data into subplots (small multiples) to compare across categories.

* **Statistics (stat)**

  * Data transformations before plotting (e.g., counts for histograms, regression lines).

* **Scale**

  * Controls mapping of data to aesthetics (linear/log scale, color gradients, axis limits).

* **Themes**

  * Overall styling (background color, fonts, grid lines).

* **Guides**

  * Elements that help interpret plots (axes, legends, labels).

* **Interactivity**

  * Making plots dynamic (hover tooltips, zooming, filtering).

* **Annotations**

  * Adding extra info: text, arrows, highlights, markers on the chart.

* **Extensions**

  * Add-on packages or custom functions to enhance visualizations beyond the core library.


# -----


💡 Think of it this way:\
👉 **Data** is *what* you plot,\
👉 **Aesthetic + Geometry** is *how* you show it,\
👉 **Facets, Scales, Themes, Guides** refine *the look & readability*,\
👉 **Statistics, Interactivity, Annotations, Extensions** make it *powerful & useful*.



---
### **🖼️ Parts of a Plot**

* **Title**

  * The main heading that explains what the chart is about.

* **Axis Labels**

  * Descriptions for the x-axis and y-axis, showing what each axis represents.

* **Tick Marks**

  * Small marks (with or without numbers) along the axes that show scale or measurement intervals.

* **Grid Lines**

  * Faint lines across the plot that make it easier to read values and align data points.

* **Legend**

  * A guide that explains symbols, colors, or line styles used to represent different data categories.

* **Data Points / Data Series**

  * The actual plotted values (bars, dots, lines, etc.) that represent your data.

* **Data Labels**

  * Numbers or text shown directly on data points or bars to highlight exact values.

* **Annotations**

  * Extra notes, arrows, or highlights added to emphasize specific parts of the plot.

* **Background**

  * The overall area behind the chart (can be styled with colors, themes, or images).


> **💡 Quick tip**\
When you make a plot, keep it **clean and simple** — don’t overload with too many labels or grid lines. The goal is to make data **easy to read** at a glance ✨.

### Import Libraries and Load Dataset

```python
# import dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset

df = sns.load_dataset('diamonds')

df.head()
```


### 💎 Columns description in Diamonds Dataset

* **carat** 
  * Weight of the diamond (bigger carat = heavier & usually more valuable).

* **cut**

  * Quality of the diamond’s cut (Fair, Good, Very Good, Premium, Ideal).
  * Affects brilliance/sparkle.
* **color**

  * Diamond’s color grade (from **D** = best/clear to **J** = more yellow).

* **clarity**

  * Measure of how clear the diamond is (few or many imperfections).
  * Ranges from “IF” (flawless) to “I3” (more visible flaws).

* **depth**

  * Total depth % of the diamond = (z / mean(x, y)) × 100.
  * Relates to proportions, affecting sparkle.

* **table**

  * Width of the diamond’s top surface (table) compared to its widest point.

* **price**

  * Price of the diamond (in US dollars).

* **x**

  * Length of the diamond (in mm).

* **y**

  * Width of the diamond (in mm).

* **z**

  * Depth/height of the diamond (in mm).

# **---**

**💡Simple way to remember:**
* **carat, cut, color, clarity** → describe **quality**.
* **depth, table, x, y, z** → describe **dimensions**.
* **price** → the **value**.

## Basic plotting with Matplotlib

```python
plt.bar(df['carat'] , df['price'])
```

**Interpratation:**

This bar plot does not provide meaningful information because both axis are numerical.

```python
plt.scatter(df['carat'] , df['price'])
plt.title("Relationship between diamon carat and price")
plt.xlabel("Carat Weight")
plt.ylabel("Price (USD)")
plt.show()
```

**Interpretation:**

* The vast majority of diamonds in the dataset are less than 1 carat in weight, with a very high concentration of data points below 0.7 carats.

**Insight:**

* The plot shows a strong positive correlation, but not a perfect one. For a fixed carat weight, there is a large variance in price. This indicates that other factors, such as cut, color, and clarity, are also major determinents of a diamond's price.

```python
plt.scatter(df['price'] , df['depth'])
plt.title('comparson of depth vs price')
plt.xlabel('price($)')
plt.ylabel('depth')
plt.show()
```

**Interpretation:**

* Depth is not a major factor in a diamond's price.



```python
plt.bar(df['cut'].unique() , df.groupby('cut')['price'].mean())
plt.title('Average Price by cut')
plt.xlabel("Cut Quality")
plt.ylabel("Average Price")
plt.show()
```

**Interpretation:**

* On the basis of cut quality , Premium quality has the highest price while Ideal quality has the lowest price.

**Insights:**

* Lowest average price of ideal quality shows that price depends upon other features than quality.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to display all available Seaborn color palettes
def display_palettes():
    palette_names = sorted(sns.palettes.SEABORN_PALETTES)
    n = len(palette_names)
    
    # Create subplots
    fig, axes = plt.subplots(nrows=n, figsize=(8, 2 * n))
    
    for ax, palette_name in zip(axes, palette_names):
        colors = sns.color_palette(palette_name)
        for i, color in enumerate(colors):
            ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=color))
        ax.set_xlim(0, len(colors))
        ax.set_ylim(0, 1)
        ax.set_title(palette_name, fontsize=10, pad=10)
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()

display_palettes()
```
